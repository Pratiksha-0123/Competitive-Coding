```python
# WAP to accept percentage and if per >90-A
# per > 80-B
# per > 60-C
# per < 60-Fail
per = int(input("Enter Percentage:"))
if per >= 90:
  print("A")
elif per >= 80 and per < 90:
  print("B")
elif per>=60 and per <80:
  print("C")
else:
  print("Fail")
```

```python
name ="prashu"
data =['a','e','i','o','u']
vowels =0
consonants = 0
for i in name:
  if i in data:
    vowels += 1
  else:
    consonants += 1
  print("vowels" ,vowels)
  print("consonants" ,consonants)
```
